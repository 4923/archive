### Create new project
> 전반적으로 메타데이터를 잘 활용할 수 있어야 한다. (잘) 구성된 HTML 레이아웃을 제공하고 그에 맞추어 렌더링되므로 적절한 메타데이터가 필요할 것.

> 효율적인 시간 관리를 위해 migration 대신 새 프로젝트를 생성했다.

**quarto 명령어로 새 프로젝트 생성**

- 일단 최소한의 구조를 만들어 준다. fastpages는 레포 클론해서 PR날리는 등 번거로운 절차가 있는 반명 이건 cli 환경에서 바로 프로젝트를 생성해주므로 편한 것으로 보인다.
- migration 과정이 아닌 단순 프로젝트 생성에도 동일한 명령어를 사용하는데 이 때의 명령어는 : `quarto create-project {project name}` 이다.

```bash
$ quarto create-project --type website:blog .   // 기존 dir
$ quarto create-project {project name}          // 새 프로젝트
$ quarto install extension quarto-ext/video
```

## Sidebar

수동으로 설정할 수 있으나 매번 설정해주기 번거로우므로 `contents: auto` 옵션을 설정한다. 자동 설정을 켜면 title에 적힌 내용을 기준으로 section이 설정되며 `{project name}/_quarto.yml` 에서 아래와 같이 설정한다.

```yml
website:
  sidebar:            # https://quarto.org/docs/websites/website-navigation.html#side-navigation 
    style: floating
    search: true
    contents: auto    # https://quarto.org/docs/websites/website-navigation.html#auto-generation
```

현재 quarto의 공식 repo에서는 auto 옵션을 채용하지 않는 것으로 보이므로 해당 옵션은 사용해보고 변경해보도록 한다.   
주의 할 점은 root path에서 auto 설정을 해 둘 경우 검색시 모든 파일이 탐색 결과에 포함된다는 것.  

> Using `contents: auto` at the root level will result in all documents in your website being included within the navigation (save for the home page which can be navigated to via the title link). Navigation is constructed using the following rules:
You can also include automatically generated items in the middle of a list of normal items by including an item with an `auto` property. Here we add an `auto` entry in the middle of a list of items: [reference](https://github.com/quarto-dev/quarto-web/blob/d1377f630e08bffa57e6ed678837ae49cc405481/docs/websites/website-navigation.qmd#L180)


### example `_quarto.yml`

```yml
# ./git lonbs/_quarto.yml

project:
  type: website

execute:
  echo: true

format:
  html:
    # https://bootswatch.com/sketchy/
    theme:
      light: Lux
      dark: Superhero
    css: styles.css
    toc: true
    
    # ====
    # CODE
    # ====
    
    # https://quarto.org/docs/output-formats/html-code.html#code-overflow
    code-overflow: wrap
    # https://quarto.org/docs/output-formats/html-code.html#code-tools
    code-fold: true   
    code-tools: true
    # https://quarto.org/docs/output-formats/html-code.html#line-numbers
    highlight-style: oblivion
    code-line-numbers: false
    code-block-border-left: true
# https://quarto.org/docs/output-formats/html-code.html#highlighting
# Quarto will automatically select the appropriate style based upon the code chunk background color’s darkness. 
# Users may always opt to specify the full name (e.g. atom-one-dark) to by pass this automatic behavior.


metadata-files: [nbdev.yml, sidebar.yml]

website:
  repo-actions: [issue]
  
  sidebar:            # https://quarto.org/docs/websites/website-navigation.html#side-navigation 
    style: floating
    search: true
    contents: auto    # https://quarto.org/docs/websites/website-navigation.html#auto-generation

  favicon: favicon.ico
  comments: 
    utterances:
      repo: 4923/archive
```

### Quick Rendering


1. [publish(recommend)] 해당 프로젝트 PATH에 이동하여 `quarto publish` 를 입력하면 gh-pages 를 통해서 바로 배포가 진행된다.
```sh
$ quarto publish
```
2. [render] quarto는 Render 를 통해 라이브서버 또는 로컬호스트로 서버 돌리지 않아도 바로 볼 수 있다는 장점이 있다.
```sh
# preview
$ quarto preview
```
```sh
# temporary preview
$ quatro render {target project}

# or Using shortcut key `Ctrl + Shift + K` renders specific file
# in those case, control (ctrl) key in Windows matches cmd when it comes to MacOS
```

### Categorizing Issue Bypass

1. Issue found 
    Github Action으로 배포시 `sidebar: contents: auto` 설정에서 categorizing에 문제가 생긴다. 
2. Solution
    CLI 명령어로 배포하게 되면 기본적으로 ([Quarto Pub](https://quarto.org/docs/publishing/); publishing service for quarto docs etc. free, easy to use) `quarto pub`가 선택되는데, destination이 자동으로 `gh-pages`로 잡혔다면 그대로 엔터 눌러서 진행해도 된다. quarto publish 명령어를 통해 배포하면 동일한 카테고리끼리 잘 묶인다.
3. 따라서, (2023.05.11 기준) quarto publish를 진행할 때에는 quarto pub를 사용하여 배포하고, git push를 통해 repo를 갱신했을 때에도 `quarto publish` 에서 quarto pub을 선택해 재배포하는 과정을 거치는 편이 좋다.
